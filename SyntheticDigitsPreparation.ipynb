```python
from PIL import Image
import os
import random
import csv

# Set paths to your digit images for different fonts
font_directories = [
    "/content/SameFont/Kaustubh-1",
    "/content/SameFont/Kaustubh-2",
    "/content/SameFont/Kaustubh-3",
    "/content/SameFont/Kaustubh-4",
    "/content/SameFont/Kaustubh-5",
    "/content/SameFont/Kaustubh-6",
    "/content/drive-download-20240107T111107Z-001/vansh-1",
    "/content/drive-download-20240107T111107Z-001/vansh-3",
    "/content/drive-download-20240107T111107Z-001/vansh-4",
    "/content/drive-download-20240107T111107Z-001/vansh-6",
    # Add more directories as needed
]

output_path = "/content/sample_data/generated_data_model1"
```

```python

# Create output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Set the number of images you want to generate for each font
num_images_to_generate_per_font = 100

# Initialize a list to store the generated data for the CSV file
csv_data = []

for font_directory in font_directories:
    # Check if the specified directory exists
    if not os.path.exists(font_directory):
        raise FileNotFoundError(f"The directory '{font_directory}' does not exist.")

    # List all digit images in the current font directory
    digit_images = os.listdir(font_directory)

    # Check if there are digit images in the directory
    if not digit_images:
        raise FileNotFoundError(f"No digit images found in the directory '{font_directory}'.")

    for index in range(num_images_to_generate_per_font):
        # Randomly select multiple digit images
        selected_digits = random.sample(digit_images, k=4)  # Adjust k as needed

        # Read the selected digit images
        digit_images_list = [Image.open(os.path.join(font_directory, digit)) for digit in selected_digits]

        # Resize the digit images if necessary
        resized_digits = [digit.resize((50, 50)) for digit in digit_images_list]

        # Create a white background
        background_width = sum(digit.width for digit in resized_digits)
        background = Image.new("RGB", (background_width, 50), "white")

        # Record the combination of digits used
        used_digits = ''.join(os.path.splitext(digit)[0] for digit in selected_digits)

        # Paste the digits onto the white background with reduced spacing
        x_position = 0
        digit_boxes = []  # List to store bounding boxes for each digit
        for digit, digit_image in zip(selected_digits, resized_digits):
            box = (x_position, 0, x_position + digit_image.width, 50)
            digit_boxes.append({"Digit": os.path.splitext(digit)[0], "Box": box})
            background.paste(digit_image, (x_position, 0),digit_image)
            x_position += digit_image.width - 10  # Adjust the spacing here

        # Save the generated image
        unique_filename = f"combined_{os.path.basename(font_directory)}_{used_digits}_{index}.png"
        output_file_path = os.path.join(output_path, unique_filename)
        background.save(output_file_path)

        # Record data for the CSV file
        csv_data.append({
            "FontName": os.path.basename(font_directory),
            "UniqueFileName": unique_filename,
            "UsedDigits": used_digits,
            "GeneratedNumber": index,
            "BoundingBoxes": digit_boxes,  # Add bounding boxes to the CSV data
        })

# Write CSV file
csv_file_path = "/content/sample_data/generated_data_model1.csv"
csv_columns = ["FontName", "UniqueFileName", "UsedDigits", "GeneratedNumber", "BoundingBoxes"]

with open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=csv_columns)
    writer.writeheader()
    for data in csv_data:
        writer.writerow(data)
```
